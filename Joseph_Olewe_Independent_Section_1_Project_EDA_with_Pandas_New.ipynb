# **Data Prep**

```python
# Dislaying dataset to see the information I want to keep
df.head()
```

```python
# Droping the 'studio' column from the dataset using copy so that code knows a new dataset and not a refrence
df = df.drop(['studio'], axis=1).copy()
```

```python
# Check to see if change has been made (columns = 4)
df.shape
```

```python
# Changing 'foreign_grossing' column from an object to a float
df['opening_data_cleaned'] = pd.to_numeric(df['foreign_gross'], errors='coerce')
```

```python
df.dtypes
```

```python
# Droping 'foreign_gross' column
df = df.drop(['foreign_gross'], axis=1).copy()
```

```python
# Renaming column
df = df.rename(columns={'opening_data_cleaned': 'Foreign Gross',
                        'title' : 'Title',
                        'domestic_gross' : 'Domestic Gross',
                        'year': 'Year' })
```

```python
df.head()
```

```python
# Merging df1 and df2 so that the movie and their rating are both displayed on a single dataset
df1 = df1.merge(df2, on='tconst', how='left')
df1.head()
```

```python
# Droping 'tconst' and 'original_title' columns
df1 = df1.drop(['tconst', 'original_title'], axis=1).copy()
```

```python
# Check to see if columns are droped
df1.columns
```

```python
df1.dtypes
```

```python
# Renaming column
df1 = df1.rename(columns={'primary_title': 'Title',
                          'start_year': 'Year',
                          'runtime_minutes':'Runtime (Minutes)',
                        'genres': 'Genres',
                        'averagerating': 'Rating',
                        'numvotes': 'Number of Votes'})
```

```python
df1.head()
```

```python
df = df.merge(df1, on='Title', how='left')
df.head()
```

```python
# Dropping repeat colunms
df.columns
```

```python
df = df.drop(['Year_y'], axis=1).copy()
```

```python
df.head()
```

```python
# Renaming column
df = df.rename(columns={'Year_x': 'Year'})
```

```python
df.head()
```

```python
# Highlight missing values
df.isna().sum()
```

```python
# Highlighting duplicate data
df.loc[df.duplicated()]
```

```python
# Delete duplicte rows and reset the index
df = df.loc[~df.duplicated(subset=['Title'])] \
       .reset_index(drop=True).copy()
```

```python
df.head()
```

```python
# What genra of movies are the highest rated?
ax = df['Genres'].value_counts()
df.query('Genres != "Documentary,Drama,Romance"') \
  .groupby('Genres')['Rating'] \
  .agg(['mean', 'count']) \
  .head(10) \
  .query('count >= 10') \
  .sort_values('mean')['mean'] \
  .plot(kind='barh', figsize=(12,5), title='Highest Rated Genres')
plt.show()
```
